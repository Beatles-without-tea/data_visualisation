## Data Visualization

```python
import seaborn as sns
```

### Neighbourhood

```python
neighbourhood_counts = airbnb['neighbourhood'].value_counts()

plt.figure(figsize=(12, 6))
neighbourhood_counts.plot.bar()

plt.xlabel('Neighbourhood')
plt.ylabel('Count')
plt.title('Neighbourhood')
plt.show()
```

### Room Type

```python
neighbourhood_counts = airbnb['room_type'].value_counts()

plt.figure(figsize=(12, 6))
neighbourhood_counts.plot.bar()

plt.xlabel('Room Type')
plt.ylabel('Count')
plt.title('Room Type')
plt.show()
```

## Map of Neighbourhood

```python
import contextily as ctx
plt.figure(figsize=(15, 10))
sns.scatterplot(x='longitude', y='latitude', hue='neighbourhood', data=airbnb,alpha=0.6, palette='Reds')
ctx.add_basemap(ax=plt.gca(), crs='EPSG:4326', source=ctx.providers.Stamen.TonerLite)

plt.legend(loc = 'upper center',bbox_to_anchor = (0.5, -0.15),ncol =3)
plt.show()
```

## Availability of Room

```python
plt.figure(figsize=(15, 8))
sns.scatterplot(x='longitude', y='latitude', hue='availability_365', data=airbnb)
ctx.add_basemap(ax=plt.gca(), crs='EPSG:4326', source=ctx.providers.Stamen.TonerLite)
plt.show()
```
